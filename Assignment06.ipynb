```python
import numpy as np
import matplotlib.pyplot as plt

num     = 201
std     = 20
a       = 2 
b       = 10

n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-100,100,num)
y1      = a * x + nn * std + b 
y2      = a * x + b

plt.plot(x, y1, 'b.', x, y2, 'r')
plt.show()

# x  : x-coordinate data
# y1 : (noisy) y-coordinate data
# y2 : (clean) y-coordinate data 
# y = f(x) = a * x + b
```

# Find $\theta=(A^TA)^{-1}A^Tb$  and $\theta=(\theta_1, \theta_2)$

```python
def straightLine(v_X,v_Y):
    m = []; m_A = []
    for i in range(0, len(x)):
        test = np.array((1, v_X[i]))
        m.append(test)
    m_A = np.mat(m)
    v_Y = np.mat(v_Y).T
    ATA = m_A.T * m_A
    if(np.linalg.det(ATA) == 0.0):
        print("This matrix is singular")
        return
    weight = ATA.I * (m_A.T * v_Y) 
    return weight
```

# $y = \theta_1 + \theta_2x$

```python
weight = straightLine(x, y1)
print(weight)
y = float(weight[1]) * x + float(weight[0])
plt.plot(x, y1, 'b', x, y, 'r')
```
